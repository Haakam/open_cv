#  my first deep learning model classificatioon

```python
import keras
import tensorflow as tf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
import seaborn as sns
```

```python
( x_train,y_train),(x_test,y_test) =keras.datasets.fashion_mnist.load_data()
```

```python
class_names=['T-shirt/top','Trouser','Pullover','Dress','Coat','Sandal','Shirt','Sneaker','Bag','Ankle boot'] #label
```

```python
j=1
for i in np.random.randint(0,1000,25):
    plt.subplot(5,5,j)
    plt.imshow(x_train[i],cmap='Greys')
    plt.title(class_names[y_train[i]])
    plt.axis('off')
    j+=1
```

```python
plt.figure(figsize=(10,10))
for i in range(25):
    plt.subplot(5,5,i+1)
    plt.imshow(x_train[i],cmap='Greys')
    plt.title(class_names[y_train[i]])
    plt.axis('off')
```

#  Feature scaling

```python
x_train=x_train/255
x_test=x_test/255 #0 to 1
```

```python
x_train[0]
```

# Building a neural Neetwork

```python
model=keras.models.Sequential([keras.layers.Flatten(input_shape=[28,28]),
                              keras.layers.Dense(units=128,activation='relu'),# 128 are neurons
                               keras.layers.Dense(units=10,activation='softmax')])# units 10 is because 10 classes
```

```python
model.summary()
```

# optimizer is an algorithm used to adjust the weights of the network to minimize the loss function.

```python
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
model.fit(x_train,y_train,epochs=1)
```

```python
model.fit(x_train,y_train,epochs=5)
```

# test and evaluate neural network

```python
model.evaluate(x_test,y_test)
```

```python
y_pred=model.predict(x_test)
```

# confusion matrix

```python
from  sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test, [np.argmax (i)for i in y_pred])
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(cm,annot=True,fmt="d")
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, [np.argmax (i)for i in y_pred]))
```
