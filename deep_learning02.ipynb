# scalling data

```python
x_train=x_train/255
x_test=x_test/255
```

```python
from sklearn.model_selection import train_test_split
x_train,x_val,y_train,y_val = train_test_split(x_train,y_train,test_size=0.2,random_state=2020)
```

# Building convolutional network

###Maxpooling:to reduce the spatial dimensions (width and height) of the input volume, which helps in reducing the computational load and preventing overfitting.

```python
model=keras.models.Sequential([keras.layers.Conv2D(filters=32,kernel_size=3,activation='relu',
                        strides=(1,1),padding='valid',input_shape=(28,28,1)),
                        keras.layers.MaxPooling2D(pool_size=(2,2)),
                        keras.layers.Flatten(),
                        keras.layers.Dense(units=128,activation='relu'),#units means neurons
                        keras.layers.Dense(units=10,activation='softmax')])
```

```python
model.summary()
```

```python
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

## we train the model

```python
model.fit(x_train,y_train,epochs=10,batch_size=512,verbose=1,validation_data=(x_val,y_val))
```

## testing model

```python
model.predict(np.expand_dims(x_test[0],axis=0)).round(2)#axis=0: This indicates that the new axis will be added at the 0th position (the beginning of the array's shape).
```

```python
np.argmax(model.predict(np.expand_dims(x_test[0],axis=0)).round(2))
```

## for cross checking

```python
y_test[0]
```

```python
y_pred=model.predict(x_test).round(2)
y_pred
```

## evaluate

```python
model.evaluate(x_test,y_test)
```

## visulize output

```python
plt.figure(figsize=(15,30))
j=1
for i in np.random.randint(0,1000,60):
    plt.subplot(10,6,j)
    plt.imshow(x_test[i].reshape(28,28),cmap=plt.cm.binary)
    plt.axis('off')
    plt.title('actual={}/{}\npredicted={}/{}'.format(class_names[y_test[i]],y_test[i],class_names[np.argmax(y_pred[i])],np.argmax(y_pred[i])))
    j=j+1
```

```python
from sklearn.metrics import confusion_matrix
plt.figure(figsize=(15,10))
y_pred_labels=[np.argmax(label) for label in y_pred]
cm=confusion_matrix(y_test,y_pred_labels)
sns.heatmap(cm,annot=True,fmt='d',xticklabels=class_names,yticklabels=class_names)
```

```python
from sklearn.metrics import classification_report
cr=classification_report(y_test,y_pred_labels,target_names=class_names)
print(cr)
```

### Cnn building complex model

```python
Cnn_model2=keras.models.Sequential([
                              keras.layers.Conv2D(filters=32,kernel_size=3,activation='relu',
                                        strides=(1,1),padding='valid',input_shape=(28,28,1)),
                               keras.layers.MaxPooling2D(pool_size=(2,2)),
                               keras.layers.Conv2D(filters=64,kernel_size=3,activation='relu',
                                        strides=(2,2),padding='same'),
                               keras.layers.MaxPooling2D(pool_size=(2,2)),
                               keras.layers.Flatten(),
                               keras.layers.Dense(units=128,activation='relu'),#units means neurons
                               keras.layers.Dropout(0.25),
                               keras.layers.Dense(units=128,activation='relu'),
                               keras.layers.Dense(units=10,activation='softmax')
                               ])
```

# compiling cnn model

```python
Cnn_model2.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
Cnn_model2.fit(x_train,y_train,epochs=20,batch_size=512,verbose=1,validation_data=(x_val,y_val))
```

```python
Cnn_model2.evaluate(x_test,y_test)
```

## very complex model

```python
Cnn_model3=keras.models.Sequential([
                              keras.layers.Conv2D(filters=64,kernel_size=3,activation='relu',
                                        strides=(1,1),padding='valid',input_shape=(28,28,1)),
                               keras.layers.MaxPooling2D(pool_size=(2,2)),
                               keras.layers.Conv2D(filters=128,kernel_size=3,activation='relu',
                                        strides=(2,2),padding='same'),
                               keras.layers.MaxPooling2D(pool_size=(2,2)),
                               keras.layers.Conv2D(filters=64,kernel_size=3,activation='relu',
                                        strides=(2,2),padding='same'),
                              keras.layers.MaxPooling2D(pool_size=(2,2)),
                               keras.layers.Conv2D(filters=64,kernel_size=3,activation='relu',
                                        strides=(2,2),padding='same'),
                               keras.layers.Flatten(),
                               keras.layers.Dense(units=128,activation='relu'),#units means neurons
                               keras.layers.Dropout(0.25),
                              keras.layers.Dense(units=256,activation='relu'),#units means neurons
                               keras.layers.Dropout(0.50),
                               keras.layers.Dense(units=128,activation='relu'),
                              keras.layers.Dense(units=256,activation='relu'),
                              keras.layers.Dropout(0.25),
                              keras.layers.Dense(units=128,activation='relu'),
                               keras.layers.Dropout(0.10),
                               keras.layers.Dense(units=10,activation='softmax')
                               ])
```

## compling model

```python
Cnn_model3.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
Cnn_model3.fit(x_train,y_train,epochs=50,batch_size=512,verbose=1,validation_data=(x_val,y_val))
```

```python
Cnn_model3.evaluate(x_test,y_test)
```
